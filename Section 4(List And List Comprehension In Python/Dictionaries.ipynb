```python
## Dictionaries are unorders collection of items. They store data in key value pairs. Key must be unique and immutable
# (eg. Strings,numbers,or tuples), while values can be of any type.
```

```python
# creating dictionarie
empty_dict = {}
print(type(empty_dict))
```

```python
# dictionarie example

dict = {"name":"Abhinav yadav",
        "age": 18,
        "proffesion":"Data Scientist"}
print(dict)
```

```python
# error which is commonly occur

std = {"name":"Abhinav yadav",
        "age": 18,
        "name":"Data Scientist"}
print(std)

# when we mention 2 key with same "name", the latest key written will be count
```

```python
# accessing the keys
print(dict)
print(dict["proffesion"])

# we can also access with "get()" method
print(dict.get("name"))

# now trying to access key which is not in the dict

print(dict.get("last_name")) # it will print none
print(dict.get("last_name","yadav"))
# So we can give any key which is not present in orginal dict a defualt value.
```

```python
# Modifying elements in dictionaries
# Dictionaries are mutable we can add,remove,and update elements in them

dict1 = {"name":"Abhinav",
        "age": 18,
        "proffesion":"Data Scientist"}
print(dict1)
```

```python
# Modifying elements

# updating,
dict1["age"] = 20
print(dict1)

# adding,
dict1["last name"]="yadav"
print(dict1)

# deleting
del dict1["last name"]
print(dict1)
```

```python
# Dictionaries method

keys = dict1.keys() # give all the kyes
print(keys)

values = dict1.values() # give all the values
print(values)

items = dict1.items() # give all the key value pairs
print(items)
```

```python
## Iterating over dictonaries
# we can use loops to iterate over keys,values,and items
print(dict1)

# iterating over keys
for keys in dict1.keys():
    print(keys)
```

```python
# iterating over values
for values in dict1.values():
    print(values)
```

```python
# iterating over items
for items in dict1.items():
    print(items)
```

```python
# same like iterating over items, i want keys and values in pair
for keys,values in dict1.items():
    print(f"{keys}: {values}")
```

```python
# Nested dictionaries

students = {
    "student1": {"name":"yash","age":32},
    "student2": {"name": "dev","age":21}
}
print(students)
```

```python
# accessing nested dictionaries
print("Name of Student 2: ",students["student2"]["name"])
print("Age of student 1: ",students["student1"]["age"])
```

```python
# iterating over nested dictionaries
print(students)

for student_id,student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key,value in student_info.items():
        print(f"{key}= {value}")
```

```python
# dictionary comperhension
dict3 = {x:x**2 for x in range(5)}
print(dict3)
```

```python
## Practicle example
# calculating the frequency of numbers repeating

numbers = [1,2,3,5,2,4,6,2,3,7,8,9,5,4,4,4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] +=1
    else:
        frequency[number] = 1
print(frequency)
```
