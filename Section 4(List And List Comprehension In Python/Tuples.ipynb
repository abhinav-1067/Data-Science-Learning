```python
## Tuples are orders  collection of items that are immutable, they are similar to list but their immutability
# make them different
```

```python
# creating a tuple
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
# secomd mehtod to creating tuple
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
# we can convert list in tuples

lst = tuple([1,2,3,4,5,6])
lst
```

```python
# and we can aslo convert tuple into list
tpl = list((1,2,3,4,5,6,7))
tpl
```

```python
# Accessing tuples elements
num = tuple((1,2,3,4,5,6))
print(num)
print(num[3])
print(num[-1])
print(num[:5])
print(num[1::2]) # same slicing method like lists
```

```python
# Tuple operations
numb = tuple((1,2,3,4,5,6))
mixed_tuple= tuple((1,3.1,"hello world",True))

concatenation_tuple = numb + mixed_tuple
print(concatenation_tuple)
```

```python
# Packing and unpcking tuple
```

```python
## Paking a tuple
packed_tuple= 1,"abhinav",True
print(packed_tuple)
```

```python
# Unpacking tuple with *
numbers = (1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested lists
## Nested Tuples
```

```python
# nested list
lts = [[1,2,3,4],['a','b','c','d'],[9,7,6,4]]
print(lts[2][0:3])
print(lts[2][0:])
```

```python
# nested tuple
tup = ((1,2,3),('a','b','c'),(True,False))
# accessing elements in nested tuples
print(tup[1][1])
print(tup[2][0:])
```

```python
## Iterating over nested tuples

for sub_tuple in tup:
    for items in sub_tuple:
        print(items,end=" ")
    print()
```
