```python
# creating sets
sets = {1,2,3,4,5}
print(sets)
print(type(sets))
```

```python
# another way to make sets
str = [1,2,3,"abhi",2.11]
tp1 = set(str)
print(tp1)
```

```python
## we can also convert list and tuples in set
lst = [1,2,3,4,5]            # list
tp = ('a','b','c','d')       # tuple

A = set(lst)
B = set(tp)
print(A)
print(B)
```

```python
## basic sets operstion

# Adding & removing
# for this we take common set
common_set = {1,2,3,4,5,6,7}
print(common_set)
```

```python
# Adding
common_set.add(8)
print(common_set)

# if we add again same element in set, then set ignore the repeated element and print it only one time.
common_set.add(8)
print(common_set)
```

```python
# removing 
common_set.add(4)
print(common_set)
common_set.remove(4)
print(common_set)

## it will remove 4 from the set
```

```python
## if we have to remove an element which is not in set then it will show key error, in that situation we use 
#  " common_set.discard(11) " so if the number is in set it will remove it or the number is not present in the set it will not show any error

common_set.discard(11)
print(common_set)
```

```python
## if i want to clear all the elements from the set

st2 = {1,2,3,4,5,6,6}
print(st2)

st2.clear()
print(st2)   # it will remove all the elements
```

```python
## to find the element is present in the set or not

st3 = {1,2,3,4,5,6,7}

print(3 in st3)
print(11 in st3)
```

```python
## Mathematical operation
set1 = {1,2,3,4,5,6,2,4}
set2 = {7,8,9,10,3,6,3}
print(set1,"\n",set2)
```

```python
# union

union_set = set1.union(set2)
print(union_set)
```

```python
# intersection
intersection_set = set1.intersection(set2)
print(intersection_set)
```

```python
# Difference
# with difference we remover common elemnts in set 1 with is common with set 2 elements

difference_set = set1.difference(set2)
print(difference_set)
```

```python
# symmetric difference

symmetric_set = set1.symmetric_difference(set2)
print(symmetric_set)
```

```python
## Sets Method
st1 = {1,2,3}
st2 = {3,4,5}
# is subset
print(st1.issubset(st2))

# is superset
print(st1.issuperset(st2))

```
