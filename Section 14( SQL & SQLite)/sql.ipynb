### SQL & SQLite

SQL ( Structured Qurey Language) is standard language for managing and manipulating relational databases.
SQLite is self-contained, serverless, zero configuration database engine that is widely used for embedded
database system.

```python
import sqlite3
```

```python
# Connect with sqlite database

connection = sqlite3.connect('example.db')
connection
```

```python
# Creating an cursor which itrate through the database

cursor = connection.cursor()
```

```python
# Creating Table

cursor.execute('''
           Create Table If Not Exists employees(
               id Integer Primary Key,
               name Text Not Null,
               age Interger,
               department Text
           )    
               ''')

# commiting changes
connection.commit()
```

```python
# Now Inserting the data in sqlite table that we create above,

cursor.execute("INSERT INTO employees VALUES (1, 'Abhinav', 18, 'Data Scientist')")

connection.commit()
```

```python
# Query the data from the table

cursor.execute("Select * from employees")
rows = cursor.fetchall()

# print the query data

for row in rows:
    print(row)
```
