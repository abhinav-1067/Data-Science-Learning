```python
# simple syntax of "defination" function
num = list(range(1,50))

def even_or_odd(num):
    if num%2==0:
        print(f"{num} number is even number")
    else:
        print(f"{num} number is odd number")
```

```python
# now when i create a function we can call it anywhere
even_or_odd(43)
```

```python
# functions with multiple parameters
def add(a,b):
    return a+b

returns = add(4,9)
print(returns)
```

```python
# default parameter
def greet(name="Guest"):
    print(f"Hello {name} welcome to our hotel")
    
greet("Abhinav")
greet()
```

```python
# variables lenght arguments
# positional and keyword arguments

def noob(*abhi):
    for x in abhi:
        print(x)

noob(1,2,3,4,5,6,7,"abhinav",3.1,True)
```

```python
# i can write anything in the place of abhi but globally everyone write "args" so always use this,
# Positionsl arguments

def noob(*args):
    for x in args:
        print(x)

noob(1,2,3,4)
```

```python
# keywords arguments,
# use two stars for key & values

def bot(**args):
    for keys,values in args.items():
        print(f"{keys}: {values}")

bot(name="abhinav",region="khargone",age="18")
```

```python
# Converting temperature in farenheit or celsius

def convert_temp(temp,unit):
    if unit=='C':
        print((temp*9/5)+32)  # celsius to farenheit
    elif unit=='F':
        print((temp-32)*5/9)  # farenheit to celsius
    else:
        print("Unit not recognised") 
        
convert_temp(288,"C")
convert_temp(53,'F')
convert_temp(64,'K')
```
