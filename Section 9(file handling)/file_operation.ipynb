```python
# reading a file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
# read file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
# read file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # .strip() removes the new line character
```
