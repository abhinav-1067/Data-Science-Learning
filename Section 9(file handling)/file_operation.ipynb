```python
# read file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
# read file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # .strip() removes the new line character
```

```python
## writing a file(overwrite)

with open('example.txt','w') as file:
    file.write("Hello abhinav\n")
    file.write("This is new line")
    
    #so it will overwrite what is already written in the file
```

```python
 # writing a file without over writing
 
with open('example.txt','a') as file:
    file.write("\nThis line adding in file without overwriting ")
```

```python
## wiritng a list of lines

lines = ['\nfirst line\n','second line\n','third line']

with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## binary file

data = b'\x00\x01\x02\x03\x04'

with open('example.bin','wb') as file:
    file.write(data)
    
## and for reading binary file use "rb" instead of "wb"
```

```python
# read the content from source text file and write in destination file
# copying text
with open('source.txt','r') as source_file:
    content = source_file.read()
# now pasting the file
with open('destination.txt','w') as desti_file:
    desti_file.write(content)
```

```python
# read the text file and count the number of line,words and character.

def count_text_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        words_count = sum(len(line.split())for line in lines)
        char_count = sum(len(line)for line in lines)
    return line_count,words_count,char_count

file_path = 'example.txt'
line_count,words_count,char_count = count_text_file(file_path)
print(f"Lines: {line_count}\nWords: {words_count}\nCharacters: {char_count}")
```
