```python
# Question 1

def generator():
    for i in range(1,5):
        yield i
gen = generator()
```

```python
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# Question 2

def even_num(n):
    for i in range(n):
        if i%2==0:
            yield i

eve = even_num(23)
```

```python
print(next(eve))
print(next(eve))
print(next(eve))
print(next(eve))
```

```python
# Question 4
class MyIter:
    def __init__(self):
        self.num = 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.num <= 3:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration
itr = MyIter()
for i in itr:
    print(i)
```
