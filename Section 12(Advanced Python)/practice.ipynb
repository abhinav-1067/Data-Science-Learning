```python
# Question 2

def even_num(n):
    for i in range(n):
        if i%2==0:
            yield i

eve = even_num(23)
```

```python
print(next(eve))
print(next(eve))
print(next(eve))
print(next(eve))
```

```python
# Question 4
class MyIter:
    def __init__(self):
        self.num = 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.num <= 3:
            val = self.num
            self.num += 1
            return val
        else:
            raise StopIteration
itr = MyIter()
for i in itr:
    print(i)
```

```python
# Question 5

def any_num():
    x = 5
    def inner(n):
        return n + x
    return inner

num = any_num()
print(num(10))
```

```python
# Question 6
def double(func):
    def wrapper(a,b):
        return func(a,b) * 2
    return wrapper

@double
def add(a,b):
    return a + b
print(add(2,3))
```
