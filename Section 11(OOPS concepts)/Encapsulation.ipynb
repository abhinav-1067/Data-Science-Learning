```python
# What it is:
#Encapsulation means wrapping data (variables) and methods (functions) together inside one class and controlling direct access to them.
```

```python
# Definition in one line:
# 👉 Encapsulation is the process of hiding data inside a class and allowing access only through methods.
```

```python
# Encapsulation with Getter and Setter method,
# public,protected,private variables
```

```python
# Public
class Person:
    def __init__(self,name,age):
        self.name = name    #public variables
        self.age = age      #public variables
        
def get_access(per):
    print(f"Persons name is {per.name}")
    
person = Person("Abhinav",32)
get_access(person)
```

```python
dir(person)

# in this "person" directory you can see name and age, which means it is public variables
```

```python
# Private
class Person:
    def __init__(self,name,age,gender):
        self.__name = name    #private variables
        self.__age = age      #private variables
        self.gender = gender  #public variable
```

```python
person = Person("abhinav",18,"Male")
dir(person)

# In this directory you can see, now name and age is not public variables, but gender is pubilc variable. Because for private variable we use double dash (__) after dot (.)
```
