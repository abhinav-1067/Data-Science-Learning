```python
# Instance variables and meathod.
class Dog:
    # constructor
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    def bark(self):
        print(f"{self.name} say bhoow")

dog1 = Dog("tommy",2)
dog1.bark()
```

```python
# Modeling a bank account
class bank:
    def __init__(self,owner,balance=0):
        self.owner = owner
        self.balance = balance
    
    def deposite(self,amount):
        self.balance += amount
        print(f"{amount} is deposited in account. New balance is {self.balance}")
    
    def withdraw(self,withd):
        if withd>self.balance:
            print("Insufficient balance!")
        else:
            self.balance -= withd
            print(f"{withd} is credit from the account. New balance is {self.balance}")
    
    def bala(self):
        return self.balance

# create an account
own1 = bank("abhinav",120000)
print(f"Owner: {own1.owner}\nBalance: {own1.balance}")
```
