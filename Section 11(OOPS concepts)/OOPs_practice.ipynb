```python
class Greeter:
    def greet(self,name):
        print(f"Hello, {name}!")

# Example of usage

gree = Greeter()
gree.greet("Shruti")
```

```python
class Calculator:
    def add_(self,a,b):
        return a+b
    def subtract_(self,a,b):
        return a-b

num = Calculator()
print("Addition: ",num.add_(4,6))
print("Difference: ",num.subtract_(2,5))
```

```python
class Employee:
    def __init__(self,name,id=None,department=None):
        self.name = name
        self.id = id
        self.department = department
    def show_details(self):
        print(f"Name: {self.name}")
        if self.id:
            print(f"ID: {self.id}")
        if self.department:
            print(f"Department: {self.department}")

e1 = Employee("Abhinav",None,"IT")
e2 = Employee("Surbhi",24345,"IT")

print(e1.show_details())
print(e2.show_details())
```
