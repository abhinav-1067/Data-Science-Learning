```python
class Greeter:
    def greet(self,name):
        print(f"Hello, {name}!")

# Example of usage

gree = Greeter()
gree.greet("Shruti")
```

```python
class Calculator:
    def add_(self,a,b):
        return a+b
    def subtract_(self,a,b):
        return a-b

num = Calculator()
print("Addition: ",num.add_(4,6))
print("Difference: ",num.subtract_(2,5))
```

```python
# Question 1
class Employee:
    def __init__(self,name,id=None,department=None):
        self.name = name
        self.id = id
        self.department = department
    def show_details(self):
        print(f"Name: {self.name}")
        if self.id:
            print(f"ID: {self.id}")
        if self.department:
            print(f"Department: {self.department}")

e1 = Employee("Abhinav",None,"IT")
e2 = Employee("Surbhi",24345,"IT")

print(e1.show_details())
print(e2.show_details())
```

```python
# Question 2

class Car:
    def __init__(self,brand,speed):
        self.brand = brand
        self.speed = speed
    def show_details(self):
        print(f"Cars Brand: {self.brand}")
        print(f"Cars Speed: {self.speed}")
car = Car("BMW",243)
car.show_details()
```

```python
# Question 4
class Dog:
    def Name(self,name):
        self.name = name
    def Bark(self,say):
        self.say = say
    def final(self):
        print(f"{self.name} says {self.say}!")
        
dog = Dog()
dog.Name("Builder")
dog.Bark("Woof")

dog.final()
```

```python
# Question 5
class BankAccount:
    def Balance(self,balance=None):
        self.balance = balance
    
    def deposite(self,amount):
        self.amount = amount
        self.balance = self.balance + self.amount
        
    def withdraw(self):
        self.balance = self.balance - self.amount
        
    def final_bal(self):
        print(f"Your Balance Is: {self.balance}")
        
bal = BankAccount()
bal.Balance(2040)
bal.deposite(60)

bal.final_bal()
```
