```python
# Polymorphism is a core concepts of object oriented programming, it allows objects of different classes to be treated as common superclass. It provides a way to perform single action in different forms. Polymorphism is typically achived through method overriding and interfaces.
```

```python
# METHOD OVERRITING

# method over allows a child class to provide a specific implementaion of method that is already defined in its parents class.
```

```python
# Base class
class Animal:
    def speak(self):
        return "how animal speak"
    
# Derived class 1
class Dog(Animal):
    def speak(self):
        return "Woof"

# Derived class 2
class Cat(Animal):
    def speak(self):
        return "Meow"

dog = Dog()
print(dog.speak())
cat = Cat()
print(cat.speak())
```

```python
# base class 1
class Shape:
    def area(self):
        return "The Shape of the figure"
    
# Derived class 1
class Rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
# Derived class 2    
class Circle:
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return 3.14*(self.radius**2)

# Funtion that demonstrate polymorphism
def ploy_area(shape):
    print(f"The area is: {shape.area()}")
    
rectangle = Rectangle(5,7)
circle = Circle(3)

ploy_area(rectangle)
ploy_area(circle)
```
