```python
# Abstraction is a methode to hide complex code and show only neccessary thing
```

```python
from abc import ABC , abstractmethod

class Vehicals(ABC):
    def usage(self):
        print("This car is used for driving")
    
    @abstractmethod
    def start_engine(self):
        pass
    
class Car(Vehicals):
    def start_engine(self):
        print("Car engine started")

def operate_vehical(vehical):
    vehical.start_engine()
    vehical.usage()

car = Car()
operate_vehical(car)
carr = Vehicals()
operate_vehical(carr)
```
