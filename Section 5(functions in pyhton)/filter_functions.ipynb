```python
# filter with lambda function
numbers = [1,2,3,4,56,7,8,9,10]
greater_than_five = list(filter(lambda x:x>5,numbers))
print(greater_than_five)
```

```python
# using filter() to check weather the age is greater than 25 or not in dictionaries

people = [
    {"name":"abhinav",'age':35},
    {'name':'dev','age':42},
    {'name':'gay','age':24}
]

def greater_than_25(person):
    return person['age']>25
psp = list(filter(greater_than_25,people))
print(psp)
```
