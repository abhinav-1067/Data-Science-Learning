```python
## lambda functions are small anonymous functions define using lambda keyword. They can have any number of argumenst but only one expression. They are used as shorts operations or as arguments of higher-order functions.
```

```python
# syntax
lambda arguments: expression
```

```python
# arguments through defination function
def addition(a,b):
    return a+b
print(addition(3,2))
```

```python
# same arguments with lambda function
addition = lambda a,b: a+b
print(addition(3,2))
```

```python
# map()-- map is used to apply function to all the items in the list

numbers = [1,2,3,4,5]
list(map(lambda x: x**2,numbers))
```
