### Logging In Pyhton

Logging in pyhton means recording message about what your programm while it runs. Instead of using print() Logging is used in real projects to track error, warning, or useful info.

Where is it used?

--Debugging errors 🐞
--Tracking program flow
--Saving logs in files for later analysis
--Used in almost every real-world application (web apps, data pipelines, etc.)

```python
import logging

#Configure the basic logging setting
logging.basicConfig(level=logging.DEBUG)

# log message with different servirety leveles

logging.debug("This is an debug message")
logging.error("This is an error message")
logging.warning("This is an warning message")
logging.info("This is an info message")
logging.critical("This is an critical message")
```

```python
## Logging Levels (important),

# DEBUG → detailed info (for developers)
# INFO → normal messages
# WARNING → something unexpected
# ERROR → serious issue
# CRITICAL → program may stop
```

```python
# we can also change the format of this logging system

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

# logging messages

logging.debug("This is an debug message")
logging.error("This is an error message")
logging.warning("This is an warning message")
logging.info("This is an info message")
logging.critical("This is an critical message")
```

```python
# we can also change the format of this logging system

import logging

logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

# logging messages

logging.debug("This is an debug message")
logging.error("This is an error message")
logging.warning("This is an warning message")
logging.info("This is an info message")
logging.critical("This is an critical message")
```
