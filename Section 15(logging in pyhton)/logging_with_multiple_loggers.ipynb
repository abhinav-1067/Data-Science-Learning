### Logging In With Multiple Loggers

Instead of using one global logger(logging), you can create seperate loggers for different parts of your programms.
For example -- 1. One logger for database, 2. One logger for authentication, 3. One logger for main app .

```python
import logging

# first create loggers
logger1 = logging.getLogger("module1")
logger2 = logging.getLogger("module2")

# now setting level
logger1.setLevel(logging.INFO)
logger2.setLevel(logging.ERROR)
```

```python
# Now pactising a real world example,

# creating logger
app_logger = logging.getLogger("app")
db_logger = logging.getLogger("database")

# setting levels
app_logger.setLevel(logging.INFO)
db_logger.setLevel(logging.ERROR)

# file handler to store messages,
file_handler = logging.FileHandler("app.log")

# now i can change the formate of loggers
formatter = logging.Formatter("%(asctime)s-%(name)s-%(date)s-%(message)s")
file_handler.setFormatter(formatter)

# now adding handler
app_logger.addHandler(file_handler)
db_logger.addHandler(file_handler)

app_logger.info("App started successfully")
db_logger.error("DataBase connection failed")
```
