```python
# Example
try:
    num = int(input("Enter the number: "))
    print(10/num)
except ValueError:
    print("You entered invalid value!!")
except ZeroDivisionError:
    print("Enter number which is greater than '0'")
```

```python
# try,except,else block
try:
    numb = int(input("Enter the number: "))
    result=10/numb
except ValueError:
    print("You enterd invalid value!! ")
except ZeroDivisionError:
    print("Enter value grater than '0'!!")
except Exception as ex3:
    print(ex3)
else:
    print(f"The result is {result}")
```

```python
# try,except,else block
try:
    numb = int(input("Enter the number: "))
    result=10/numb
except ValueError:
    print("You enterd invalid value!! ")
except ZeroDivisionError:
    print("Enter value grater than '0'!!")
except Exception as ex3:
    print(ex3)
else:
    print(f"The result is {result}")
finally:
    print("Execution is completed!")
    
# when try,except,or block else is executed 'finally' block is also executed.
```

```python
# exception handling and file handling

try:
    file = open("example3.txt",'r')
    content = file.read()
    print(content)
except FileNotFoundError:
    print("File not found!")
finally:
    if 'file' in locals() and not file.closed:
        file.close()
        print('file close')
```
