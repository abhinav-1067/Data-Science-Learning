```python
## for loops
for i in range(5):
    print(i)
```

```python
## when we inverse the range
for i in range(10,1,-1):     # we wright skipable number in negative sign

    print(i)             
```

```python
## While loop
# in while loop it executes the code as log as the condition is true.

count = 0
while count<5:
    print(count)
    count=count+2 #it is very important to write condition
```

```python
## Loop Control Statements

# Break Statement
for i in range(10):
    if i==6:
        break
    print(i)
```

```python
# Continue Statement
# condition mai jo value or kuch bhi nikale continue statement usse skip kardega

for i in range(10):
    if i%2==0:
        continue
    print(i)
    
```

```python
# Pass Statement
# the pass statement is an null operation, it does nothing,

for i in range(5):
    if i==3:
        print("this number is",i)
        pass
    print(i)
```

```python
#Example
## sum of first ten N natural numbers.

n=10
sum=0
count=1

while count<=n:
    sum = sum+count
    count = count+2
print("Sum of first 10 even numbers are: ",sum)
```
