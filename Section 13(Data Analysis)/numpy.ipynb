```python
# create a array

import numpy as np

arr1 = np.array([10,20,30])
print(arr1)
print(type(arr1))
print(arr1.shape)
```

```python
# reshaping an array

arr2 = np.array([1,4,2,5,5])
print(arr2.reshape(1,5))
```

```python
# Numpy vectorised operations
arr3 = np.array([1,2,3,4,5])
arr4 = np.array([10,20,30,40,50])

# element wise addition,
print(f"Addition: {arr3+arr4}")

# element wise substraction,
print(f"Substraction: {arr3-arr4}")

# element wise multiplication,
print(f"Multiplication: {arr3*arr4}")
```

```python
# Universal functions
arr = np.array([2,4,3,5,7])

# square root,
print(np.sqrt(arr))

# exponential,
print(np.exp(arr))

# Sine,
print(np.sin(arr))

# natural log,
print(np.log(arr))
```
