### Medium Level --Pandas-- Questions

```python
import numpy as np
```

```python
# Question 1

data3 = {
    "Name": ["Aman", "Riya", "Sohan", "Neha"],
    "Marks": [45, 82, 67, 30]
}

df2 = pd.DataFrame(data3)
df2
```

```python
df2['Result'] = np.where(df2['Marks'] >= 50 , 'Pass', 'Fail')
df2
```

```python
# Question 2

dt = {
    "Name": ["Aman", "Riya", "Sohan"],
    "City": ["bhopal", "delhi", "bhopal"]
}

df3 = pd.DataFrame(dt)
df3
```

```python
df3['City'] = df3['City'].str.title()
df3
```

```python
# Question 3

data = {
    "Name": ["Aman", "Riya", "Sohan"],
    "Marks": [80, 95, 90]
}

df = pd.DataFrame(data)
df
```

```python
top_student = df.loc[df['Marks'].idxmax(),'Name']
print(top_student)
```

```python
# Question 4

data = {
    "Marks": [35, 60, 75, 20, 90]
}

df = pd.DataFrame(data)
df
```

```python
failed_stu = df[df['Marks'] < 40]
print(len(failed_stu))
```

```python
# Question 5

data = {
    "Price": [100, 200, 300],
    "Quantity": [2, 1, 4]
}

df = pd.DataFrame(data)
df
```
