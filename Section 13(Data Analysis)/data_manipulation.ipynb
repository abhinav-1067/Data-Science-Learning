```python
# Merging and joining dataframes

df1 = pd.DataFrame({'Keys':['A',"B",'C'],'Values1':[1,2,3]})
df2 = pd.DataFrame({'Keys':['A',"E",'B'],'Values2':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
# Merfing dataframe with Key colunm
pd.merge(df1,df2,on="Keys",how="inner")
```

```python
# Merfing dataframe with Key colunm
pd.merge(df1,df2,on="Keys",how="outer")
```
