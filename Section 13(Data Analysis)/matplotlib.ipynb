```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,15,25]

plt.plot(x,y)
plt.xlabel('x-aixs')
plt.ylabel('y-aixs')
plt.title('Basic Graph Plotting')
plt.show()
```

```python
# Creating an custom line,

x = [1,2,3,4,5]
y = [1,4,9,15,25]

plt.plot(x,y, color='red', linestyle='--',marker='o')
plt.grid(True)
plt.show()
```

```python
# Bar Plot

catergories = ['A','B','C','D','E','F']
values = [1,6,3,4,5,7]

plt.bar(catergories,values,color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
# scatter plot

x = [1,2,3,4,5]
y = [2,4,7,6,3]

plt.scatter(x,y,color='black',marker='x')
plt.show()
```

```python
# Pie Chart
labels = ['TMKB','BSDK','BKL','TMB']
sizes = [30,20,28,22]
colors = ['gold','lightskyblue','yellowgreen','lightcoral']

plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('OG Galis')
plt.show()
```

```python
# Pie Chart
labels = ['TMKB','BSDK','BKL','TMB']
sizes = [30,20,28,22]
colors = ['gold','lightskyblue','yellowgreen','lightcoral']
explode = [0.2,0,0,0]

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.title('OG Galis')
plt.show()
```

```python
# bar graph plotting with real data
import pandas as pd
data = pd.read_csv('sales_data.csv')
data
```

```python
# plotting a bar graph of product with respect to its revenue

total_revenue = data.groupby('Product')['Revenue'].sum()
total_revenue
```

```python
total_revenue.plot(kind='bar',color='blue')
```

```python
sales_data = data.groupby('Date')['Revenue'].sum().reset_index()
plt.plot(sales_data['Date'],sales_data['Revenue'])
plt.show()
```
