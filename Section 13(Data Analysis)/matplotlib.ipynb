```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,15,25]

plt.plot(x,y)
plt.xlabel('x-aixs')
plt.ylabel('y-aixs')
plt.title('Basic Graph Plotting')
plt.show()
```

```python
# Creating an custom line,

x = [1,2,3,4,5]
y = [1,4,9,15,25]

plt.plot(x,y, color='red', linestyle='--',marker='o')
plt.grid(True)
plt.show()
```
