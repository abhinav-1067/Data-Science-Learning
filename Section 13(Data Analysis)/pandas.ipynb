Pandas--- DataFrame & Series

Pandas is a powerful data manipulation library in pyhton, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A series is one-dimensional array like object, while a dataframe is two dimensional, size mutable and potentially heterogeneous tabular data structure while with labeled axes.

```python
# Series

import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series\n",series)
```

```python
# Series with dictonaries
data1 = {'a':1,'b':2,'c':3}
ser = pd.Series(data1)
print(ser)
```
